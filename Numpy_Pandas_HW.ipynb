#Numpy

```python
# Create a NumPy array with values from 10 to 49 (inclusive).
#● Reshape it into a 4x10 matrix.
#● Then extract the last column from the reshaped matrix

import numpy as np

arr1=np.random.randint(10,50, size=(40))
print(arr1)
arr1=arr1.reshape(4,10)
print("Matrix of 4*10:\n",arr1)
print("last column:\n",arr1[0:,-1])
```

```python
#2. Create two 3×3 matrices:
#One filled with all 5s , One with values from 1 to 9
#● Add the two matrices
#● Multiply them element-wise
#● Then find the transpose of the result

A = np.full((3, 3), 5)
B = np.random.randint(1,10, size=(3,3))
print(A)
print(B)
C=np.add(A,B)
print(C)
D=np.multiply(A,B)
print(np.transpose(C))
print(np.transpose(D))
```

```python
#3. Create a NumPy array of 15 random integers between 10 and 100.
#● Replace all values greater than 50 with -1
#● Count how many values were replaced.

arr2=np.random.randint(10,101, size=(20))
print(arr2)
x=np.where(arr2>50)
print(x)
for i in x:
  arr2[i]=-1
print(arr2)
```

```python
#4. Generate a 5×4 array with random values between 0 and 1 (float).
#● Find the row with the maximum sum of values.
#● Normalize each row using the formula:

mat1=np.random.random(size=(5,4))
arr3=np.sum(mat1, axis=1)
print(mat1)
print(arr3)
print(np.max(arr3))
max1=mat1.max(axis=1, keepdims=True)  # axis=1 takes the row maximum and keepdims keeps the max of each matrix a matrix
min1=mat1.min(axis=1, keepdims=True)
normalised=(mat1-min1)/(max1-min1)
print(normalised)
```

```python
#5. arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
#● Find all unique elements
#● Find the duplicates (elements that appear more than once)

arr4 = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique, count=np.unique(arr4, return_counts=True)
print(count)
l=np.array([], dtype=np.int64)
for i in range(len(count)):
  if count[i]>1:
    l=np.append(l,unique[i])
print(l)

```

```python
#6. Create a 6×6 NumPy array with values from 1 to 36.
#● Extract the middle 4×4 submatrix.
#● Then extract all even numbers from this submatrix.

mat2 = np.random.randint(1,37, size=(6,6))
print(mat2)
mat2_sub=mat2[1:5,1:5]
print(mat2_sub)
even_nums = mat2_sub[mat2_sub % 2 == 0]
print(even_nums)
```

```python
#7. Generate an array of shape (3, 4) with values from 0 to 11.
#● Reshape it to (2, 6)
#● Then flatten it into a 1D array
#● Convert the 1D array back to shape (4, 3)

mat3 = np.random.randint(1,12, size=(3,4))
mat4 = mat3.reshape(2,6)
print(mat3)
print(mat4)
mat5 = mat4.reshape(12)
print(mat5)
mat6=mat5.reshape(4,3)
print(mat6)
```

```python
#8 . Create an array: a = np.array([[1, 2], [3, 4]])
#● Append a new row [5, 6]
#● Then append a new column [7, 8, 9]

mat7 = np.array([[1, 2], [3, 4]])
mat7 = np.append(mat7, [[5, 6]], axis=0)
print(mat7)
mat8 = np.append(mat7, [[7], [8], [9]], axis=1)
print(mat8)
```

```python
#9.a = np.array([1, 2, 3])
#b = np.array([4, 5, 6])
#c = np.array([7, 8, 9])
#● Join them into a single array
#● Reshape the result into a 3×3 matrix
#● Extract the diagonal elements

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])

arr5=np.concatenate((a,b,c))
print(arr5)
mat9=arr5.reshape(3,3)
print(mat9)
diag_elements = np.diag(mat9)
print(diag_elements)
```

```python
#10. Create a NumPy array of integers from 1 to 20.
#● Extract all values that are divisible by 3 or 5.
#● Count how many such values exist.

arr6 = np.random.randint(1,12, size=(20))
values = arr6[(arr6 % 3 == 0) | (arr6 % 5 ==0)]  #we cannot use or in array
print(arr6)
print(values)
print(len(values))
```

#Pandas

```python
#1. Load the Titanic dataset and display the first 10 rows.
import seaborn as sns
data = sns.load_dataset('titanic')
data.head(10)
```

```python
#2. Find the number of missing values in each column.
data.isna().sum()
```

```python
#3. Find the percentage of passengers who survived, grouped by gender.
data.groupby('sex')['survived'].mean()*100
```

```python
#4. Find the average, median, and standard deviation of passenger ages.

print(data['age'].mean())
print(data['age'].median())
print(data['age'].std())
```

```python
#Count how many passengers were in each class (pclass).
data['class'].value_counts()
```

```python
#6. Show all passengers who were female, under age 18, and survived.
data[(data['sex']=='female') & (data['age']<18) & (data['alive']=='yes')]
```

```python
#7. Replace missing age values with the median age.
data["age"].fillna(data["age"].median())
```

```python
#8. Create a column age_group with values:
#a. 'Child' for age < 18
#b. 'Adult' for age 18–60
#c. 'Senior' for age > 60
#d. Use: apply() or np.where() or custom function

def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

data['age_group'] = data['age'].apply(categorize_age)

```

```python
#9. Display the top 5 passengers who paid the highest fare.
top5 = data.sort_values(by='fare', ascending=False).head(5)
print(top5)
```

```python
#10. Find the average fare paid, grouped by both pclass and sex.
data.groupby(['class', 'sex'])['fare'].mean()
```

```python
#11. Rename the following columns:
#a. 'sex' → 'gender'
#b. 'fare' → 'ticket_fare'

data.rename(columns={'sex': 'Gender', 'fare': 'Ticket_Fare'}, inplace=True)
print(data)
```

```python
#12. Count the number of passengers who survived and didn’t survive, grouped by class.

survival_counts = data.groupby(['class', 'alive']).size()
print(survival_counts)
```

```python
#13. Find the average age of passengers grouped by both class and gender.

data.groupby(['class', 'Gender'])['age'].mean()
```
